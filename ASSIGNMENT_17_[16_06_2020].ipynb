```python
import numpy as np
import pandas as pd
import sklearn
```

```python
data3=pd.read_csv("https://data.princeton.edu/wws509/datasets/salary.dat",sep=" ")
data3
```

```python
#Check for any missing values in the dataset
data3.apply(lambda x:x.isnull().sum(),axis=0)
```

```python
data3.dtypes
```

```python
#Display the data of only the non-numerical columns
data3.select_dtypes(include=['object'])
```

```python
#Display the data of only the numerical columns
data3.select_dtypes(include=['float64'])
```

```python
#Get a list of categorical variables with the categories frequency count.
data3.apply(pd.value_counts)
```

```python
#Create dummy variables for the columns sx and dg
data4=pd.get_dummies(data3.sx)
data4
```

```python
data5=pd.get_dummies(data3.dg)
data5
```

```python
#Create a dataframe by adding dummy variables to the original data frame using concat.
df=pd.concat([data3,data4],axis=1)
df
```

```python
df1=pd.concat([df,data5],axis=1)
df1
```

```python
#Create another  dataframe by adding dummy variables to the original data frame using merge
data1=pd.merge(df1,data3)
data1
```

```python
#After adding the dummy columns to the dataframe, drop the original columns sx and dg
df1.drop(['sx','dg'], axis = 1) 
```

```python
#Also, drop one of the dummy columns created for sx and dg variables
df1.drop(['doctorate','15'], axis = 1) 
```
